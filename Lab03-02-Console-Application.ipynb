## 2. Create Application

การทดลองนี้ จะสร้างโปรเจ็กต์ .NET console application ในชื่อ "HelloWorld"

1. เปิด Visual studio code
2. เรียก command palette โดยการกด Ctrl+Shift+P หรือเรียกจากเมนู View->Command Palette...
3. ใน command palette พิมพ์ .NET new project แล้วกด enter



4. รอสักครู่
5. พิมพ์ console app แล้วกด enter




6. ตั้งชื่อโปรเจค  HelloWorld-Lab03  แล้วกด enter



7. เลือกที่ตั้ง project (ถ้าไม่ต้องการเปลี่ยนให้กด enter ได้เลย)



8. กำหนด option ของ project (ถ้าไม่ต้องการเปลี่ยนให้กด enter ได้เลย)




9. .net จะสร้าง console application project ตามที่กำหนด โดยจะมีโครงสร้างโฟลเดอร์ดังในรูป 




10. ให้คลิกที่ไฟล์ Program.cs จะเห็นเนื้อหาของโปรแกรมดังรูป
