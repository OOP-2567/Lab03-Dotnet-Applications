## 1. Introduction

Windows form Application เป็นเทคโนโลยีที่ใช้มายาวนาน (อาจจะมากกว่า 20 ปี) จึงมี library ด้านกราฟฟิกส์ที่ยึดโยงกับ win32 เป็นอย่างมาก

กรณีดังกล่าวจึงเป็นเหตุผลหนึ่งที่ทำให้มีความยากในการออกแบบ windows form บน editor อื่นๆ ที่ไม่ใช่ Visual Studio

เพื่อไม่ให้นักศึกษาต้องเรียนรู้การใช้งานเครื่องมือพัฒนาที่หลากหลานและเกินจำเป็น การทดลองในวิชานี้ จะมีขอบเขตอยู่ใน Visual Studio Code เป็นหลัก 

สำหรับผู้ที่ต้องการสร้าง windows form application ที่ใช้งานจริง ต้องไปเรียนรู้การใช้ visual studio ซึงก็ไม่ยากในการทำความเข้าใจ

https://learn.microsoft.com/en-us/dotnet/desktop/winforms/?view=netdesktop-8.0


## 2. Create Application

1. เปิด visual studio code 

__หมายเหตุ__
   เนื่องจากไม่มีตัวเลือกในการสร้าง .net project จากเมนู View->Command Palette... จึงต้องใช้ terminal ในการสร้างและทดสอบโปรเจค

2. เปิด terminal โดยใช้เมนู view->terminal หรือกด Ctrl + Shift + ` 

   2.1 ตรวจสอบว่าเป็น folder ที่จะสร้าง project ไว้ในนั้นจริง ๆ  

3. พิมพ์ `dotnet new winforms -n HelloForm1 -o HelloForm1` ใน terminal
4. พิมพ์คำสั่ง `dir` ใน terminal เพื่อดูว่ามี folder ชื่อ HelloForm1 ที่ถูกสร้างขึ้นใหม่อยู่แล้วหรือยัง ถ้ายังไม่มี อาจจะเกิดจากการสร้างไม่สมบูรณ์
5. พิมพ์คำสั่ง `cd HelloForm1` เพื่อเปลี่ยน folder  ที่ทำงานไปอยู่ในโปรเจค HelloForm1

## 3. Build Application

6. พิมพ์คำสั่ง `dotnet build`  ใน terminal เพื่อ build โปรแกรม

## 4. Run Application
7. พิมพ์คำสั่ง `dotnet run` ใน terminal เพื่อรันโปรแกรม

15 วางเป้าหมายสำหรับโปรแกรมที่จะสร้าง



โปรแกรมที่จะสร้าง ประกอบด้วยช่องข้อความสำหรับใส่ตัวเลข 2 ตัว เป็นตัวตั้งและตัวบวกตามลดับ

ถัดลงมาเป็นปุ่มกดสำหรับการบวกเลชสองจำนวนเข้าด้วยกัน

ด้านล่างสุด เป็นช่องข้อความที่ใส่ผลลัพธ์ที่ได้จากการบวก

ด้านซ้ายของช่องข้อความจะมีชื่อตัวแปรกำกับไว้ทั้งสามตัว

16.1 เพิ่ม code สำหรับการแสดงผลบน windows form โดยจะเพิ่มต่อเนื่องลงไปด้านล่างตามลำดับ

ให้เพิ่ม code สำหรับการสร้าง object ที่จะวางบน windows form  ไว้ถัดลงมาจากบรรทัด   `this.Text = "Form1";`




 Code ที่ต้องเพิ่ม


>    ```cs
>    // สร้าง object ของ controls ที่จะวางบน form
>    buttonAdd = new Button();
>    label_A = new Label();
>    label_B= new Label();
>    label_C= new Label();
>    textBox_A = new TextBox();
>    textBox_B = new TextBox();
>    textBox_C = new TextBox();
>
>    // หยุดการแสดงผลบน form ชั่วคราว (freeze) โดยคำสั่ง SuspendLayout(); 
>    // เพื่อไม่ให้เกิดการกระพริบในระหว่างที่เพิ่ม control ทีละตัว
>    // กรณีดังกล่าวอาจจะเกิดขึ้นได้ในตอนที่คอมพิวเตอร์ทำงานหนัก หรือการ์ดแสดงผลทำงานช้า
>    // ซึ่งจะทำให้เกิดความไม่สบายตาแก่ผู้ใช้
>    // controls ต่างๆ ที่ถูกเพิ่มหลังคำสั่ง SuspendLayout(); จะไม่ถูกนำไปแสดงผลบนหน้าจอ
>    // ต้องใช้คำสั่ง ResumeLayout เพื่อยกเลิกการ freez หน้าจอ
>    SuspendLayout();
>
>    // เพิ่ม label สำหรับตัวแปร A
>    label_A.AutoSize = true;
>    label_A.Location = new Point(25, 24);
>    label_A.Name = "label_A";
>    label_A.Size = new Size(15, 15);
>    label_A.Text = "A";
>    Controls.Add(label_A);
>
>
>    // เมื่อเพิ่ม controls เสร็จแล้ว ให้เรียก ResumeLayout(false); เพื่อ update หน้าจอชอง form
>    // จะทำให้โปรแกรมสามารถรันได้ตามปกติ
>    ResumeLayout(false);
>    PerformLayout();
>    ```
